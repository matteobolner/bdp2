This notebook contains the python code necessary to import the data describing the progression of COVID-19 cases in italy, provided on github by the Protezione Civile department (https://github.com/pcm-dpc/COVID-19)

```python
#import the libraries necessary to access the data from the URL, handle the JSON format and plot the data
import json
import requests
import matplotlib.pyplot as plt
```

```python
source = requests.get('https://raw.githubusercontent.com/pcm-dpc/COVID-19/master/dati-json/dpc-covid19-ita-andamento-nazionale.json')
```

```python
covid_db = source.json()
```

```python
#extract the data obtained and group it in a dictionary
db_dict = {}
for name in covid_db[1]:
    db_dict[name]=[]
for i in covid_db:
    for j in i.items():
        db_dict[j[0]].append(j[1])
```

```python
db_dict.keys()
```

```python
#plot date and ICU occupation numbers
#y axis can be substituted with whatever other variable is available in db_dict
plt.plot(db_dict['data'], db_dict['terapia_intensiva'])
plt.xticks([])
plt.xlabel('Date')
plt.ylabel('Patients in ICU')
```
